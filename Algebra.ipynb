```python
a1=np.array([2,-5])
a2=np.array([-1,3])
x=np.array([1,-4])

#проверим линейную независимость векторов
def check_lin_dep(v1,v2):
    A=np.array([v1,v2]).T
    rank = np.linalg.matrix_rank(A)
    print('vectors are linearly independent' if rank==A.shape[1] else 'vectors are linearly dependent')
check_lin_dep(a1,a2)
```

```python
#найдем координаты вектора х в новом базисе
# [x](B) = B (inv) * [X]

def coord_new_basis(v,B1,B2):
    B=np.array([B1,B2]).T
    B_inv=np.linalg.inv(B)
    x_B=np.dot(B_inv,v)
    return x_B
n=coord_new_basis(x,a1,a2)
n
```

```python
#найдем координаты вектора y в стандартном базисе
# [x] = B * [X](B)

def coord_st_basis(y_B,B1,B2):
    B=np.array([B1,B2]).T
    y_e=np.dot(B,y_B)
    return y_e
yB=np.array([1,1])
coord_st_basis(yB,a1,a2)
```

### Problem 4

```python
def low_rank_appr_err(n):
    A=np.random.rand(n,n)
    # SVD 
    U, s, VT = np.linalg.svd(A)
    errors=[]
    for r in range(2,n+1):
        Sigma = np.diag(s)
        mat_approx = U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]
        diff=A-mat_approx
        errors.append(np.linalg.norm(diff, ord='fro'))
    return errors
```

```python
ranks=range(2,101)
E=low_rank_appr_err(100)
sns.scatterplot(ranks,E)
plt.xlabel('Rank');
plt.ylabel('||A-A$_($$_a$$_p$$_p$$_r$$_)$||$_F$');
```
